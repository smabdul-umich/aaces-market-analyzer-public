# Params exploration

Browse the aircraft catalog and try parameter overrides. All data lives in `params.py` — this notebook only imports it.

```python
from params import AircraftFamily, AircraftSelection, AirlineSelection, demand

aircrafts = AircraftSelection()
print("All aircraft:", aircrafts.names)
print()
print("Sustainable concept family:")
for name in aircrafts.names:
    if aircrafts.family(name) == AircraftFamily.NOVEL_AIRCRAFT_CONCEPT:
        print(f"  {name:12s}  casm={aircrafts.casm(name):.3f}  price={aircrafts.price(name)}  seats={aircrafts.seats(name)}")
```

```python
# Apply a temporary override (does not modify params.py on disk)
aircrafts.apply_mofd({"Concept_A": {"casm": 0.10, "price_adjustment": -2}})
print("Concept_A casm (adjusted):", aircrafts.casm("Concept_A", apply_adjustments=True))
print("Concept_A price (adjusted):", aircrafts.price("Concept_A", apply_adjustments=True))
```

```python
airlines = AirlineSelection()
print("Leader budget:", airlines.budget("Leader"))
print("Study airline budget:", airlines.budget("Study1_Airline"))
print("Demand segments:", [s.name for s in demand.segments])
```
